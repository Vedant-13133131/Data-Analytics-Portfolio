```python
##Finding correlation across features
corr = client.corr()
```

```python
##Visualising the correlation
f,ax = plt.subplots(figsize=(20, 15))
sns.heatmap(corr, annot=True, fmt= '.1f',ax=ax, cmap="BrBG")
sns.set(font_scale=1.25)
plt.title("Heatmap for Client Data",fontdict={"fontsize":20})
plt.savefig("Corr.png")
```

While we do not see a lot of strong correlation acress variables, few observations are worth noting here:
##### Current consumption shows a 100% correlation with the forecasted consumption for the next calendar year. While a positive correlation is expected, a complete 100% correlation seems uncanny. Interestingly, the forecast for the following 12 months shows a less extreme 60% correlation.
##### Consumption of Gas is showing strong positive correlation with consumption of electricity 
##### The forcasted consumption is highly correlated with the net margin, although it doesn't show strong correlation with previous consumption trends
##### Max power consumption shows positive correlation with margin. Perhaps there are economies of scale which are influencing the profitability here

Since the client table has only got unique values for clients we can check their distribution

```python
plt.figure(figsize=(10, 10))
chart = sns.countplot(client['churn'])

plt.xlabel('Churn Category')
plt.ylabel('No. of Customers')
plt.title('Churned vs Non-Churned Distribution')
for p in chart.patches:
    chart.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=12, color='black', xytext=(0, 5),
                textcoords='offset points')
plt.savefig("Churned.png")
```

###### While churning may be a concerning trend for PowerCo., the loss hasn't reached a level where it cannot be recovered

```python
plt.figure(figsize=(15, 10))
chart = sns.countplot(client['channel_sales'])

plt.xlabel('Sales Channel')
plt.ylabel('No. of Customers')
plt.title('Channel Distribution')
chart.set_xticklabels(chart.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.tight_layout
for p in chart.patches:
    chart.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=12, color='black', xytext=(0, 5),
                textcoords='offset points')
```

###### For a large number of customers, it seems there is no record of the channel whereby they were connected to PowerCo. as visible from the high count of 'Missing' channel values

## Price Sensitivity

```python
####Merging Churn with the Price table
price_merged = pd.merge(price,client[['id','churn']],on="id",how = 'inner')
```

```python
price_merged
```

```python
##Visualising the correlation again to determine if current prices have any influence on churn
f,ax = plt.subplots(figsize=(20, 15))
sns.heatmap(price_merged.corr(), annot=True, fmt= '.3f',ax=ax, cmap="hot")
sns.set(font_scale=1.25)
plt.title("Heatmap for Price Sensitivity",fontdict={"fontsize":20})
plt.savefig("Price.png")
```

#### Preliminary EDA suggests that price doesn't affect customers decision to churn. While a more advanced analysis is required, which would include predictive modeling, these results should be noted
